## Things to discuss

### Here is a high-level list of human goals (*Ladan Collected the following list*):

**Personal Growth:** This category of goals includes self-improvement, personal development, and the pursuit of knowledge and new experiences. Personal growth goals might include things like learning a new skill, traveling to a new place, or taking up a new hobby.

**Career and Financial Success:** Many people set goals related to their careers or financial success, such as getting a promotion, starting a business, or saving for retirement.

**Relationships:** Goals related to relationships might include finding a romantic partner, developing deeper connections with friends or family members, or improving communication skills.

**Health and Wellness:** This category of goals includes goals related to physical, mental, and emotional well-being, such as exercising regularly, eating a healthy diet, practicing meditation, or seeking therapy.

**Contribution to Society:** Many people strive to make a positive impact on the world around them, whether through volunteering, philanthropy, or activism.


## Visualizations

The following code is a sample visualization of the above categories. Do we think we can work with them?

```python
import matplotlib.pyplot as plt

# Make data: I have 3 groups and 7 subgroups
group_names=['Personal Growth', 'Career and Financial Success', 'Relationships',
            'Health and Wellness', 'Contribution to Society']
group_size=[9,30,12,12,4]
subgroup_names=['A.1', 'A.2', 'A.3', 'B.1', 'B.2', 'C.1', 'C.2', 'C.3', 'D.1', 'D.2', 'D.3', 'D.4',
               'E.1', 'E.2', 'E.3']
# subgroup_names=['Self-improvement', 'Hobbies', 'Learning', 'Work', 'Saving-for-Retirement',
#                 'Partner', 'Family and Friends', 'Networking', 'Exercise', 'Healthy Diet', 'Mindfulness', 'Therapy',
#                'Volunteering', 'Activism', 'Philantropy']
subgroup_size=[3,3,3,24,6,6,4,2,6,3,2,1,2,1,1]

# Create colors
a, b, c, d, e=[plt.cm.spring, plt.cm.Reds, plt.cm.autumn, plt.cm.cool, plt.cm.summer]

# First Ring (outside)
fig, ax = plt.subplots(figsize=(15,8))
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=1.3, labels=group_names, colors=
[a(0.4), b(0.6), c(0.8), d(0.8), e(0.6)] )
plt.setp( mypie, width=0.3, edgecolor='white')

# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=1.3-0.3,
labels=subgroup_names, labeldistance=0.7, colors=[a(0.5), a(0.4), a(0.3), b(0.5), b(0.4),
                                                  c(0.8), c(0.7), c(0.6), d(0.6), d(0.5),
                                                  d(0.4), d(0.3), e(0.6), e(0.5), e(0.4)])
plt.setp( mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)

subgroup_names_legs=['Personal Growth','Career & Financial Success','Relationships','Health & Wellness',
                     'Contribution to Society','A.1:Self-improvement', 'A.2:Hobbies', 'A.3:Learning',
                     'B.1:Work', 'B.2:Saving-for-Retirement', 'C.1:Partner', 'C.2:Family and Friends',
                     'C.3:Networking', 'D.1:Exercise', 'D.2:Healthy Diet', 'D.3:Mindfulness', 'D.4:Therapy',
                     'E.1:Volunteering', 'E.2:Activism', 'E.3:Philantropy']
# subgroup_names_legs=['A.1:a1desc', 'A.2:a2desc', 'A.3:a3desc',
# 'B.1:b1desc', 'B.2:b2desc', 'C.1:c1desc', 'C.2:c2desc', 'C.3:c3desc',
# 'C.4:c4desc', 'C.5:c5desc']
plt.legend(subgroup_names_legs,loc='lower left')

# show it
plt.show()
```
